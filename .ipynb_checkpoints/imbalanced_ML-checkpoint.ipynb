```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from mlxtend.evaluate import bias_variance_decomp
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
import category_encoders as ce
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import roc_auc_score
import scikitplot as skplt
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_learning_curves
```

```python
df = pd.read_csv('hcvdat0.csv',usecols=['Category','Age','Sex','ALB','ALP','ALT','AST','BIL','CHE','CHOL','CREA','GGT','PROT'])
df
```

```python
#check if it's balanced
df['Category'].value_counts()
```

```python
df.isna().sum()
```

```python
# impute missing values 
df.ALB.fillna(df.ALB.mean(),inplace=True)
df.ALP.fillna(df.ALP.mean(),inplace=True)
df.ALT.fillna(df.ALT.mean(),inplace=True)
df.CHOL.fillna(df.CHOL.mean(),inplace=True)
df.PROT.fillna(df.PROT.mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df['Category'] = df['Category'].str.replace('0s=suspect Blood Donor', '4')
df
```

```python
df['Category'] = df['Category'].str.replace(r'\D', '')
df
```

```python
df['Category'].value_counts()
```

```python

OHE = ce.OneHotEncoder(cols=['Sex'],use_cat_names=True)
df = OHE.fit_transform(df)
```

```python
df=df.drop(['Sex_f'],axis=1)
```

```python
X= df.iloc[:,1:]
y= df.iloc[:,0]
```

```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE 

```

```python
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
print('Resampled dataset shape %s' % Counter(y_res))
```

Classification:
    --------

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
log_clf = LogisticRegression(solver="liblinear", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=10, random_state=42)
svm_clf = SVC(gamma="auto", random_state=42)
ran_clf = RandomForestClassifier(n_estimators=1000, n_jobs=-1, max_depth=50,
                             random_state=0)
knn_clf = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf),('ran', ran_clf),('knn', knn_clf)],
    voting='hard')
```

```python
voting_clf.fit(X_train, y_train)
```

```python
log_clf = LogisticRegression(solver="liblinear", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=10, random_state=42)
svm_clf = SVC(gamma="auto", probability=True, random_state=42)
ran_clf = RandomForestClassifier(n_estimators=1000, n_jobs=-1, max_depth=50,
                             random_state=0)
knn_clf = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='soft')
voting_clf.fit(X_train, y_train)
```

```python

for clf in (log_clf, rnd_clf, svm_clf,ran_clf,knn_clf, voting_clf, ):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

```python

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
ax = sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
```

```python
prob=svm_clf.predict_proba(X_test)
```

```python

skplt.metrics.plot_roc_curve(y_test, prob)
plt.show()
```

```python
#the learning curve
svclassifier = SVC()
plot_learning_curves(X_train, y_train, X_test, y_test, svclassifier)
plt.show()
```
